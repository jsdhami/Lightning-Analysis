```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://github.com/jsdhami/Lightning-Analysis/raw/refs/heads/main/Tri-Chandra/lightning_data_one_day.csv')
df.head()
```

```python
 fig, ax = plt.subplots()
 ax.plot(df['Time'], df['Channel A Last (mV)'])
 ax.set_xlabel('Time')
 ax.set_ylabel('Channel A Last (mV)')
 ax.set_xticks([])
 ax.grid(True)
 plt.show()
```

```python
from scipy.signal import find_peaks
# show peaks
peaks, _ = find_peaks(df['Channel A Last (mV)'], height=50, threshold=50)
print(len(peaks))


fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
ax.plot(df['Time'], df['Channel A Last (mV)']-50)
ax.axhline(y=0, color='r', linestyle='-')
ax.scatter(df['Time'][peaks], df['Channel A Last (mV)'][peaks]-50, color='red')
ax.set_xlabel('Time')
ax.set_ylabel('Channel A Last (mV)')
ax.set_xticks([])
ax.grid(True)
plt.show()
```
